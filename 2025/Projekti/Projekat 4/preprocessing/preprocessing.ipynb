## Vizualizacija distribucije podataka

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
ax = sns.countplot(
    x='Class', 
    data=df, 
    hue='Class', 
    palette={0: "green", 1: "red"},  
    legend=False  
)
plt.title('Distribucija klasa (0 = legitimno, 1 = prevara)', fontsize=14, pad=20)

for p in ax.patches:
    ax.annotate(
        f'{p.get_height():,}', 
        (p.get_x() + p.get_width() / 2., p.get_height()), 
        ha='center', 
        va='center', 
        xytext=(0, 10), 
        textcoords='offset points',
        fontsize=12
    )


plt.xlabel('Klasa', fontsize=12)
plt.ylabel('Broj transakcija', fontsize=12)
plt.xticks([0, 1], ['Legitimno (0)', 'Prijevara (1)'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

counts = df['Class'].value_counts()
percentages = counts / counts.sum() * 100

plt.figtext(
    0.5, 
    0.01,
    f"Legitimno (0): {counts[0]:,} transakcija ({percentages[0]:.2f}%) | Prijevara (1): {counts[1]:,} transakcija ({percentages[1]:.2f}%)",
    ha="center",
    fontsize=12,
    bbox={"facecolor": "white", "alpha": 0.8, "pad": 5}
)

plt.tight_layout()
plt.show()
```

## Outlieri

```python
sns.boxplot(x=df['Amount'])
plt.show()
```

## Analiza outliera

```python
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Amount'] < Q1 - 1.5 * IQR) | (df['Amount'] > Q3 + 1.5 * IQR)]
print(f"Broj outliera u 'Amount': {len(outliers)}")
```

## Skaliranje 'Amount' i 'Time'

```python
from sklearn.preprocessing import StandardScaler

df['Amount_scaled'] = StandardScaler().fit_transform(df[['Amount']])
df['Time_scaled'] = StandardScaler().fit_transform(df[['Time']])
df = df.drop(['Amount', 'Time'], axis=1)
df.head()
```
