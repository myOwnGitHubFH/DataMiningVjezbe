```python
# IQR
# inter quantile range
```

```python
lf_srednja_vrijednost = np.mean(df['LotFrontage'])
lf_std = np.std(df['LotFrontage'])
```

```python
print(lf_srednja_vrijednost)
print(lf_std)
```

```python
donja_granica_IQR = lf_srednja_vrijednost - 3*lf_std
gornja_granica_IQR = lf_srednja_vrijednost + 3*lf_std

df = df[
    (df['LotFrontage'] > donja_granica_IQR) &
    (df['LotFrontage'] < gornja_granica_IQR) # |
]
# && 
# ||
# select * from df where LotFrontage > donja_granica_IQR and LotFrontage < gornja_granica_IQR
```
