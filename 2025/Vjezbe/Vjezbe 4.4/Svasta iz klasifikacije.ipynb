```python
iris = pd.read_csv("iris.csv")
```

Vidimo da je plava klasa uvijek jasno separabilna, pa ce vjerovatno i njene metrike biti dobre...

```python
X = iris[['sepal_length','sepal_width',	'petal_length',	'petal_width']]
y = iris['species']

X, y = shuffle(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
clf = svm.SVC() # classifier
clf.fit(X_train, y_train) #SVC SUPPORT VECTOR CLASSIFIER
```

```python
y_predicted = clf.predict(X_test)
```
