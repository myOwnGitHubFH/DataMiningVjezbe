```python
df_2013 = df[df['Year'] == 2013]
df_2014 = df[df['Year'] == 2014]
```

```python
list_2013 = (df_2013['Country'].value_counts().index)
list_2014 = (df_2014['Country'].value_counts().index)
```

```python
print(len(list_2013))
print(len(list_2014))
```

```python
set_2013 = set(list_2013)
set_2014 = set(list_2014)

diff_2013_14 = set_2013.difference(set_2014)
```

```python
diff_2013_14
```

```python
df = df[~df['Country'].isin(diff_2013_14)]
```
