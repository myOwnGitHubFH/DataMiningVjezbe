```python
import pandas as pd
import matplotlib.pyplot as plt

# import express.plotly
```

# Pokusaji dokazivanja linearne separabilnost

```python
df = pd.read_csv("../Vjezbe_April_2024/datasets/LogReg/weatherAUS.csv")
```

```python
df.dropna(inplace=True)
```

```python
df = df.sample(frac = 1)
```

```python
df.head()
```

```python
X_svm = df[["MinTemp", "MaxTemp", "Pressure9am"]]
```

```python
X_svm.head()
```

```python
y_svm = df['RainTomorrow']
```

```python
y_svm.head()
```

```python
y_svm.value_counts()
```

```python
# Map 'y' to colors
color_map = {'No': 'red', 'Yes': 'green'}  # Red for 'No', green for 'Yes'
colors = df['RainTomorrow'].map(color_map)
```

```python
cols = X_svm.columns
```

```python
cols = list(cols)
```

```python
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        
        x_axis_name = cols[i]
        y_axis_name = cols[j]
        
        x_axis = X_svm[cols[i]]
        y_axis = X_svm[cols[j]]

        # Plotting
        plt.figure(figsize=(8, 6))
        plt.scatter(x_axis, y_axis, c=colors, label=f'{x_axis_name} vs {y_axis_name}')
        plt.xlabel(f'{x_axis_name}')
        plt.ylabel(f'{y_axis_name}')
        plt.legend()
        plt.grid(True)
        plt.show()
```
